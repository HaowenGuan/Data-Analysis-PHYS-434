```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
import random
import math
from scipy import stats
plt.rcParams["figure.figsize"] = (15,10)
```

### Problem 0

```python
x = np.linspace(0, np.pi*100, np.int(1e5))
y = 10+np.sin(x)
bkgd = stats.norm.rvs(size=np.int(1e5))*np.sqrt(y)+y
```

### Problem 1
**1a)**

```python
xx = x[0:1000]
slicing = bkgd[0:1000]
plt.scatter(xx, slicing)
```

**1b)**

```python
def subplot(start: int, point: int):
    xx = x[start:start + point]
    slicing = bkgd[start:start + point]
    plt.scatter(xx, slicing)
subplot(500, 2000)
```

**1c)**

```python
xx = x[0::100]
slicing = bkgd[0::100]
plt.scatter(xx, slicing)
```
