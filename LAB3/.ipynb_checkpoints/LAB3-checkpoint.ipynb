```python
d = np.append(stats.norm.rvs(loc = 12., scale = 0.4, size = 100000), [10., 10.3, 2.1, 0., 0., 15.6, 22.3, 12.7])
fig, ax = plt.subplots(1, 1)
ax.hist(d,100, density=True)
plt.tick_params(labelsize = 24)
plt.yscale('log')
#x = np.linspace(4.95,5.05,1000)
#ax.plot(x,stats.norm.pdf(x,loc = 5., scale = 0.01),linewidth = 8,alpha = 0.7)
plt.show()
```

<br> **A)**
<br> **1)** By look at data, there are appearently exist some anomaly data exist, distributed on the both side of the distribution.
<br> **2)** I believe the data that are beyond $+-5\sigma$ away from the mean are bad data.
<br> **3)** Since the standard diviation （$\sigma$）is 0.4, and the mean is 12, so $+-5\sigma$ correspond to $12 + 0.4*5 = 14$ and $12 - 0.4*5 = 10$. Therefore, for data that are smaller than 10 and bigger than 14 are bad data.
<br> **4)** Count the good and bad data as follow.

```python
BadT = 0
for x in d:
    if x < 10 or x > 14:
        BadT += 1
TrueBadT = 0
for x in [10., 10.3, 2.1, 0., 0., 15.6, 22.3, 12.7]:
    if x < 10 or x > 14:
        TrueBadT += 1
FalseBadT = BadT - TrueBadT
FalseGoodT = 8 - TrueBadT
TrueGoodT = d.size - BadT - FalseGoodT
print(TrueGoodT)
print(FalseGoodT)
print(FalseBadT)
print(TrueBadT)
```

## Problem 3
<br> **1)** If there is no moon, than we expect the background to have the pdf() of $60*8*15=7200$ convolved (summed) poisson distribution. The stronger the moon shadow, the less cosmic rays we expect to detect. So what is the probability of the number of cosmic rays we see is <= 6800?
<br> **2)** Find the pdf() of back ground by convolving (summing) a poisson distribtion with mean 1 with itself for 7200 times. Then integrate $$ \int \limits _{-\infty} ^{6800} pdf(x)dx $$

```python
d1 = stats.poisson.rvs(mu = 1, size = 10000)
for i in range(1,7200):
    d1 +=stats.poisson.rvs(mu = 1, size = 10000)
fig, ax = plt.subplots(1,2)
fig.set_size_inches(20,7)
ax[0].hist(d1, 100)
ax[0].set_title("Sum poisson with mean 1 7200 times", fontsize=24)
ax[0].set_yscale("log")
ax[0].tick_params(labelsize = 24)
ax[0].set_xlabel("x", fontsize=24)
ax[0].set_ylabel("Count", fontsize=24)

d2 = stats.poisson.rvs(mu = 7200, size = 10000)
ax[1].hist(d2, 100)
ax[1].set_title("Poisson with mean 7200", fontsize=24)
ax[1].set_yscale("log")
ax[1].tick_params(labelsize = 24)
ax[1].set_xlabel("x", fontsize=24)
ax[1].set_ylabel("Count", fontsize=24)
```

<br> I calculate and plot the semi-log graph of summing poisson with mean 1 7200 times and the poisson distribution with mean 7200 above. I found that two distribution is acutally same. So, I would simply use a poisson distribution with mean 7200.
<br> **3)** Calculation as below:

```python
d2 = stats.poisson.rvs(mu = 7200, size = 10000000)
Count = 0
for x in d1:
    if x <= 6800:
        Count += 1
prob = Count/10000000
print(prob)
print(stats.norm.ppf(prob))
```
